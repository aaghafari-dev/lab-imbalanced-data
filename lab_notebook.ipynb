```python
# Data handling and visualisation
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Preprocessing and model selection
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Regression models
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Evaluation metrics
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import cross_val_score

from sklearn.ensemble import VotingRegressor
from sklearn.neighbors import KNeighborsRegressor 
from sklearn.ensemble import StackingRegressor
from sklearn.svm import SVR
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from imblearn.over_sampling import SMOTE
from imblearn.under_sampling import TomekLinks
```

```python
# load the dataset
df = pd.read_csv('./files_for_lab/customer_churn.csv')
df.head()
    
```

```python
# Basic info about features
print("=== Dataset Info ===")
df.info()

print("\n=== Descriptive Statistics ===")
display(df.describe())
```

```python
# missing values
print("\n=== Missing Values ===")
print(df.isnull().sum())
```

```python
# Step 2: Select features and target as per instructions
features = ['tenure', 'SeniorCitizen', 'MonthlyCharges']
X = df[features]
y = df['Churn']   # target variable

# Encode target (Churn: Yes=1, No=0)
y = y.map({'Yes': 1, 'No': 0})

# Scale the independent variables (only 'tenure' and 'MonthlyCharges' are numeric; 'SeniorCitizen' is already 0/1)
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)   # scales all three, but it's fine

# Check the shapes
print("Features shape:", X_scaled.shape)
print("Target shape:", y.shape)
print("\nClass distribution:\n", y.value_counts())
```

```python
# Split data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

# Train logistic regression
lr = LogisticRegression(random_state=42)
lr.fit(X_train, y_train)

# Predictions
y_pred = lr.predict(X_test)

# Evaluation
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

**Why does a simple model give >70% accuracy?**

The dataset is imbalanced: the majority class (non‑churn, "No") makes up about 73–74% of the samples.  
A dummy classifier that always predicts "No" would already achieve ~73% accuracy.  
Therefore, even a simple logistic regression can exceed 70% by mainly predicting the majority class, without necessarily learning meaningful patterns.  
Accuracy alone is misleading; we need to look at precision, recall, and especially the F1‑score for the minority class (churn).

```python
# Apply SMOTE only on the training set 
smote = SMOTE(random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)

# Check new class distribution
print("After SMOTE:\n", y_train_smote.value_counts())

# Train logistic regression on the balanced data
lr_smote = LogisticRegression(random_state=42)
lr_smote.fit(X_train_smote, y_train_smote)

# Predict on the original test set
y_pred_smote = lr_smote.predict(X_test)

# Evaluate
print("Accuracy after SMOTE:", accuracy_score(y_test, y_pred_smote))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_smote))
print("\nClassification Report:\n", classification_report(y_test, y_pred_smote))
```

```python
# Apply TomekLinks only on the training set
tl = TomekLinks()
X_train_tl, y_train_tl = tl.fit_resample(X_train, y_train)

# Check size reduction
print(f"Original training set size: {len(X_train)}")
print(f"After Tomek Links: {len(X_train_tl)}")

# Train logistic regression
lr_tl = LogisticRegression(random_state=42)
lr_tl.fit(X_train_tl, y_train_tl)

# Predict on the same test set
y_pred_tl = lr_tl.predict(X_test)

# Evaluate
print("Accuracy after Tomek Links:", accuracy_score(y_test, y_pred_tl))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_tl))
print("\nClassification Report:\n", classification_report(y_test, y_pred_tl))
```

```python
# Quick comparison
results = pd.DataFrame({
    'Model': ['Original', 'SMOTE', 'Tomek Links'],
    'Accuracy': [accuracy_score(y_test, y_pred),
                 accuracy_score(y_test, y_pred_smote),
                 accuracy_score(y_test, y_pred_tl)],
    'Recall (minority)': [classification_report(y_test, y_pred, output_dict=True)['1']['recall'],
                          classification_report(y_test, y_pred_smote, output_dict=True)['1']['recall'],
                          classification_report(y_test, y_pred_tl, output_dict=True)['1']['recall']],
    'F1 (minority)': [classification_report(y_test, y_pred, output_dict=True)['1']['f1-score'],
                      classification_report(y_test, y_pred_smote, output_dict=True)['1']['f1-score'],
                      classification_report(y_test, y_pred_tl, output_dict=True)['1']['f1-score']]
})
print(results)
```
